**Escape characters : working with special characters within a string**\
In Python, strings are usually written between quotation marks (' ', or " "). But sometimes, you might need to include special characters inside a stringâ€”like quotation marks, backslashes, or new lines. These charcters have a special meaning in Pyhton, so if you use them notmally, Python can get confused. An example of an "illegal" character is a double quote inside a string which is already started and ended with double quotes. Here is a line from Monty Python and the Holy Grail film to illustrate:

````python
print("But if he was dying, he wouldn't bother to carve /"Aargh"/, he would just say it.")
````

This statement would generate a syntax error because Python thinks the second double quote is the end of the string and doesn't expect the rest of the text. To fix this, Python uses the escape character, which is written as a backlash [\\], followed by a letter or symbol. This tells Python, "I want to include this special character in my string - not treat it like parte of the code." Using the escape character, the statement now looks like this:

````python
print("But if he was dying, he wouldn't bother to carve \"Aargh\", he would just say it.")
````
