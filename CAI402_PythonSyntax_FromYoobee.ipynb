**Updating variables**\
It is possible to change the value of a varialbe at any time by directly assigning it a new value, or by using it in calculations.

```python
# Python example 1:

score = 5
score = 10
print(score) #displays "10" as this is the current value if "score"
```

```python
# Python example 2:

x = 5
y = 10
total = x + y
print(total) # displays "15" which is the new calculated value of "total"
```

```python
# Python example 3:

first_name = "Sam"
greeting = "Hello, " + first_name
print(greeting) # displays "Hello, Sam"
```
